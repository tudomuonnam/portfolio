```python
import pandas as pd
import matplotlib as mpl
import ffn
mpl.use('agg',warn=False)
%pylab inline
```

```python
def get_stock(sticker):
    #sticker = 'SSI'
    start = '2017-01-01'
    file = "data/" + sticker + ".csv" 
    df = pd.read_csv(file)
    df['date'] = pd.to_datetime(df['date'])
    df.index = df['date']
    del df['date']
    df = df.loc[start:]
    return df
```

```python
prices = get_stock('adj_close')
prices
```

```python
ax = prices.rebase().plot()
```

```python
returns = prices.to_returns().dropna()
ax = returns.hist(figsize(10, 5))
```

```python
stats = prices.calc_stats()
stats.display()
```

```python
ax = stats.prices.to_drawdown_series().plot()
```
