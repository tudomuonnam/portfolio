```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
start = '2016-01-01'
end = '2020-10-01'
```

```python
tickers = ['VNINDEX','REE','HCM','PNJ','VCB']
df = (pd.concat(
    [pd.read_csv(f'data/{ticker}.csv',index_col='date',parse_dates=True)['adj_close'].rename(ticker) for ticker in tickers]
    ,axis=1,sort=True))[start:].dropna()
```

```python
market = df['VNINDEX']
portf = df.loc[[df.columns != 'VNINDEX']]
```

```python
returns = df.pct_change().dropna()
returns.plot()
```

```python
print(stats.describe(returns)
```

```python
mean = np.mean(returns)
std = np.std(returns)
min = np.min(returns)
max = np.max(returns)
```

```python
x = np.linspace(min,max,100)
plt.plot(x,norm.pdf(x,mean,std))
```

```python
plt.hist(returns, edgecolor='black', density=True, bins=10)
```
